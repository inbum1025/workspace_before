```python
import matplotlib.pyplot as plt
import numpy as np
import math


r1 = 1
r2 = np.linspace(0,70,80)
x = r2/r1
u = 1 #TU/DU

print(x)

result1 = []

for r2 in np.linspace(0,70,80):
    a = (r1 + r2)/2
    et = -u/(2*a)
    v1 = math.sqrt(2*((u/r1) + et)) - math.sqrt(u/r1)
    v2 = math.sqrt(u/r2) - math.sqrt(2*((u/r2) + et))
    v0 = abs(v1) + abs(v2)
    result1.append(v0)
    


result2 = []
result3 = []
result4 = []
result5 = []

for r2 in np.linspace(0,70,80):
    rb = 50
    a1 = (r1 + rb)/2
    a2 = (rb + r2)/2
    et1 = -u/(2*a1)
    et2 = -u/(2*a2)
    v1 = math.sqrt(2*((u/r1) + et1)) - math.sqrt(u/r1)
    v2 = math.sqrt(2*((u/rb) + et2)) - math.sqrt(2*((u/rb) + et1))
    v3 = math.sqrt(u/r2) - math.sqrt(2*((u/r2) + et2))
    v4 = abs(v1) + abs(v2) + abs(v3)
    result2.append(v4)
    


for r2 in np.linspace(0,70,80):
    rb = 1000
    a1 = (r1 + rb)/2
    a2 = (rb + r2)/2
    et1 = -u/(2*a1)
    et2 = -u/(2*a2)
    v1 = math.sqrt(2*((u/r1) + et1)) - math.sqrt(u/r1)
    v2 = math.sqrt(2*((u/rb) + et2)) - math.sqrt(2*((u/rb) + et1))
    v3 = math.sqrt(u/r2) - math.sqrt(2*((u/r2) + et2))
    v4 = abs(v1) + abs(v2) + abs(v3)
    result3.append(v4)
    


for r2 in np.linspace(0,70,80):
    rb = 1000
    a1 = (r1 + rb)/2
    a2 = (rb + r2)/2
    et1 = -u/(2*a1)
    et2 = -u/(2*a2)
    v1 = math.sqrt(2*((u/r1) + et1)) - math.sqrt(u/r1)
    v2 = math.sqrt(2*((u/rb) + et2)) - math.sqrt(2*((u/rb) + et1))
    v3 = math.sqrt(u/r2) - math.sqrt(2*((u/r2) + et2))
    v4 = abs(v1) + abs(v2) + abs(v3)
    result4.append(v4)


for r2 in np.linspace(0,70,80):
    rb = 15.58
    a1 = (r1 + rb)/2
    a2 = (rb + r2)/2
    et1 = -u/(2*a1)
    et2 = -u/(2*a2)
    v1 = math.sqrt(2*((u/r1) + et1)) - math.sqrt(u/r1)
    v2 = math.sqrt(2*((u/rb) + et2)) - math.sqrt(2*((u/rb) + et1))
    v3 = math.sqrt(u/r2) - math.sqrt(2*((u/r2) + et2))
    v4 = abs(v1) + abs(v2) + abs(v3)
    result5.append(v4)



y = result1
y2 = result2
y3 = result3
y4 = result4 
y5 = result5

font1 = {'size':12}

fig = plt.figure(1,(10,6))
plt.plot(x, y,'r-')
plt.plot(x, y2, 'g-')
plt.plot(x,y3, 'b-')
plt.plot(x,y4, 'y-')
plt.plot(x,y5, color = 'limegreen')

plt.title('Hohmann vs Bi-elliptic')
plt.xlabel('r2/r1')
plt.ylabel('DEL V')
plt.vlines(11.94,0,10, color = 'gray',linestyles='solid')
plt.axis([0,70,0,1])

plt.text(13,0.8,'11.94')
plt.text(52,0.65,'r* = 15.58')
plt.text(52,0.53,'r* = 50')
plt.text(52,0.45,'r* = 200')
plt.text(52,0.4,'r* = 1000')
plt.text(4,0.25,'Hohmann', fontdict = font1)
plt.text(25,0.65,'Bi-elliptic')

plt.show()
```
